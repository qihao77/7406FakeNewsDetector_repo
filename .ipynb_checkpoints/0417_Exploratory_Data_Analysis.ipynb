# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
import gensim
```

```python
fake = pd.read_csv('Fake.csv')
```

```python
real = pd.read_csv('True.csv')
```

### Question 1: How many fake and real news?

```python
#fake is 1, real is 0
fake['class'] = 1
real['class'] = 0
```

```python
data = pd.concat([fake, real], axis=0)
```

```python
# Removing the duplicates and preventing problems with indexing
data = data.drop_duplicates()
data = data.reset_index(drop=True)
```

```python
# Display the head of data (the first beginning 5 rows)
data.head(5)
```

```python
fig, ax = plt.subplots(figsize=(4.5,5))
plt.bar(['Fake', 'Real'], data['class'].value_counts(), width=0.4, color='#4B7D95', align='center')
#ax.set_facecolor('#D1D8E2')
for i, v in enumerate(data['class'].value_counts()):
    #print(i, v)
    ax.text(i - 0.11, v + 200,str(v), color='black', fontweight='bold')
ax.set_ylabel('News Counts',size=15)
plt.tight_layout()
plt.show()
```

### Question 2: When are the news published?

**Data cleaning & correction**

```python
# Correcting some data
data['date'] = data['date'].replace(['19-Feb-18'],'February 19, 2018')
data['date'] = data['date'].replace(['18-Feb-18'],'February 18, 2018')
data['date'] = data['date'].replace(['17-Feb-18'],'February 17, 2018')
data['date'] = data['date'].replace(['16-Feb-18'],'February 16, 2018')
data['date'] = data['date'].replace(['15-Feb-18'],'February 15, 2018')
data['date'] = data['date'].replace(['14-Feb-18'],'February 14, 2018')
data['date'] = data['date'].replace(['13-Feb-18'],'February 13, 2018')


data['date'] = data['date'].str.replace('Dec ', 'December ')
data['date'] = data['date'].str.replace('Nov ', 'November ')
data['date'] = data['date'].str.replace('Oct ', 'October ')
data['date'] = data['date'].str.replace('Sep ', 'September ')
data['date'] = data['date'].str.replace('Aug ', 'August ')
data['date'] = data['date'].str.replace('Jul ', 'July ')
data['date'] = data['date'].str.replace('Jun ', 'June ')
data['date'] = data['date'].str.replace('Apr ', 'April ')
data['date'] = data['date'].str.replace('Mar ', 'March ')
data['date'] = data['date'].str.replace('Feb ', 'February ')
data['date'] = data['date'].str.replace('Jan ', 'January ')
```

**Convert the column date to datetime format**

```python
data['date'] = data['date'].str.replace(' ', '')
```

```python
for i, val in enumerate(data['date']):
    data['date'].iloc[i] = pd.to_datetime(data['date'].iloc[i], format='%B%d,%Y', errors='coerce') 

data['date'] = data['date'].astype('datetime64[ns]')
```

**Get the date range of all the news pulished**

```python
print('All news were published from ', min(data['date']).strftime('%Y-%m-%d'),\
     ' to ', max(data['date']).strftime('%Y-%m-%d'))
```

### Question 8: Word Cloud Visualization (title)

```python
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2 and token not in stop_words:
            result.append(token)
            
    return result
```

#### (1) For Fake News

```python
fake_title = data[data['class'] == 1]['title'].apply(preprocess)
fake_cloud = WordCloud(stopwords = stop_words).generate(" ".join(fake_title.apply(lambda x:" ".join(x))))
```

```python
plt.figure(figsize=(12,12))
plt.imshow(fake_cloud, interpolation = 'bilinear')
```

#### (2) For Real News

```python
real_title = data[data['class'] == 0]['title'].apply(preprocess)
real_cloud = WordCloud(stopwords = stop_words).generate(" ".join(real_title.apply(lambda x:" ".join(x))))
```

```python
plt.figure(figsize=(12,12))
plt.imshow(real_cloud, interpolation = 'bilinear')
```
